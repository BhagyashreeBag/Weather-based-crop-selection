```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.model_selection import train_test_split
from keras.preprocessing.sequence import TimeseriesGenerator
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import tensorflow as tf

mpl.rcParams['figure.figsize']=(10,8)
mpl.rcParams['axes.grid'] = False
```

```python
df=pd.read_csv('/content/temperature_prediction_data.csv')
```

```python
df
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'],format= "%d-%m-%Y")
```

```python
df.set_index('Date')[['temperature','humidity','rainfall']].plot(subplots=True)
```

```python
df_hum = df[['humidity','temperature','rainfall']]
```

```python
df_hum.describe()
```

```python
scaler = MinMaxScaler()
scaler.fit(df_hum)
data_scaled = scaler.transform(df_hum)
```

```python
data_scaled
```

```python
features = data_scaled
target= data_scaled[:,0]
```

```python
TimeseriesGenerator(features,target,length=30,sampling_rate=1, batch_size=1)[0]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(features,target,test_size=0.20,random_state=123, shuffle = False)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
win_length=60
batch_size=30
num_features=3
train_generator = TimeseriesGenerator(X_train,Y_train,length=win_length,sampling_rate=1, batch_size=batch_size)
test_generator = TimeseriesGenerator(X_test,Y_test,length=win_length,sampling_rate=1, batch_size=batch_size)
```

```python
train_generator[0]
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(128, input_shape=(win_length,num_features),return_sequences=True))
model.add(tf.keras.layers.LeakyReLU(alpha=0.5))
model.add(tf.keras.layers.LSTM(128,return_sequences=True))
model.add(tf.keras.layers.LeakyReLU(alpha=0.5))
model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.LSTM(64,return_sequences=False))
model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.Dense(1))
```

```python
model.summary()
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor ='val_loss',patience=2,mode='min')
model.compile(loss=tf.losses.MeanSquaredError(),
              optimizer=tf.optimizers.Adam(),
              metrics=[tf.metrics.MeanAbsoluteError()])
history = model.fit_generator(train_generator, epochs=50,
                              validation_data = test_generator,
                              shuffle = False,
                              callbacks =[early_stopping])
```

```python
model.evaluate_generator(test_generator,verbose=0)
```

```python
predictions=model.predict_generator(test_generator)
```

```python
predictions.shape[0]
```

```python
predictions
```

```python
Y_test
```

```python
X_test
```

```python
X_test[:,1:][win_length:]
```

```python
df_pred=pd.concat([pd.DataFrame(predictions),pd.DataFrame(X_test[:,1:][win_length:])],axis=1)
```

```python
df_pred
```

```python
rev_trans = scaler.inverse_transform(df_pred)
```

```python
rev_trans
```

```python
df_finalhum = df_hum[predictions.shape[0]*-1:]
```

```python
df_finalhum.count()
```

```python
df_finalhum['humidity_pred']=rev_trans[:,0]
```

```python
df_finalhum
```

```python
df_finalhum[['humidity','humidity_pred']].plot()
```
